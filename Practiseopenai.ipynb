```python
##Text Commpletion/Text Generation
response=openai.Completion.create(
    model="text-davinci-003",
    prompt="Act as an AI assistance explain machine learning to 6 years old children",
    max_tokens=15,
    temperature=1,
    n=5
)
print(response)
```

```python
##Text Commpletion/Text Generation(tweets prompt)
response=openai.Completion.create(
    model="text-davinci-003",
    prompt="""Act as an AI model to detect positive or negative tweets:
    Let's be honest, the only thing we want to see falling is fuel prices
    """,
    max_tokens=15,
    temperature=1,
    n=1
)
print(response)
```

```python
##Text Commpletion/Text Generation(code prompt)
prompt_msg="""Act as an AI code to explain what is wrong with code:
    code:print(HelloWorld)

    """,
response=openai.Completion.create(
    model="text-davinci-003",
    max_tokens=24,
    prompt=prompt_msg,
    temperature=1,
    n=1
)
print(response)
```
